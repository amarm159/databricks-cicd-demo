```python
#Recursion -function call itself
def sum_recusrion(arr):
    if len(arr)==1:
        return arr[0]
    else:
        return arr[0] + sum_recusrion(arr[1:])

list= [1,2,3,4,5]
print(sum_recusrion(list))    
```

```python
from pyspark.sql.functions import explode, col,explode_outer,ArrayType
from pyspark.sql.types import StructType
df = spark.read.option("multiline", "true").json(
    "/Volumes/practice_catalog/practice_schema/practice_volume"
)
# display(df)
branches = []
schema = df.schema
print(schema.fields)
def get_all_columns_from_schema(df,schema):
    branches= []
    def inner_get(df,schema,ancestor=None):
        if ancestor is None: ancestor = []
        for field in schema.fields:
            branch_path = ancestor+[field.name]
            if isinstance(field.dataType,StructType):
                inner_get(df,field.dataType,branch_path)  
            elif isinstance(field.dataType,ArrayType):
                df = df.withColumn(field.name,explode_outer(col(field.name)))    
            else:
                branches.append(branch_path)
    inner_get(df,schema)
    select_cols = []
    for item in branches:
        if len(item)>1:
            select_cols.append(col(".".join(item)).alias(".".join(item)))
        else:
            select_cols.append(col(".".join(item)))      
    display(df.select(*select_cols))       

get_all_columns_from_schema(df,schema)
# df=get_all_columns_from_schema(df,schema)
# dis

 
# display(df.select(*select_cols))        
```
